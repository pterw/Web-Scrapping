```
import statistics
from bs4 import BeautifulSoup as bs4
import requests
import csv
import pandas as pd
import numpy as np


pages_to_scrape=5
pages=[]
cars=[]
prices=[]

for i in range(1,pages_to_scrape+1):
  url = 'https://www.kijiji.ca/b-cars-trucks/ottawa/used/page-{}/c174l1700185a49?ad=offering'.format(i)
  pages.append(url)

for item in pages:
  page = requests.get(item)
  soup = bs4(page.text,'html.parser')

  for l in soup.findAll('div', class_='info'):
    if l.find(class_='dealer-logo'):
      continue
    if l.find(class_='dealer-name'):
      continue
    name = l.find(class_='title').get_text().strip()
    price = l.find(class_='price').get_text().strip()
    cars.append((name, price))

  for k in soup.findAll('div', class_='price'):
    if k.findAll(class_='dealer-logo'):
      continue
    else: 
      pricee=k.getText()
      prices.append(pricee.strip())

new_prices=[]

for i in prices:
  if i not in ('Swap / Trade','Please Contact'):
    new_prices.append(i)
new_prices = [float(i.strip(" $\n ").replace(",", "")) for i in new_prices]
print(statistics.median(new_prices))


data=pd.DataFrame(cars, columns=['name', 'price'])
data.index+=1
data
```
